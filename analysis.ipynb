```python
import numpy as np
```

```python
dataset= np.load("/Users/yumengluo/Desktop/claire/Columbia/6156/6156-imposters2/gemini_results.npz")
g_val,g_scaled,true_matrix,g_label = dataset['g_val'],dataset['g_scaled'],dataset['true'],dataset['g_label']
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.set_title("Gemini Similarity Results")
im = ax.imshow(g_val,cmap='viridis')
fig, ax = plt.subplots()
ax.set_title("True Label")
im = ax.imshow(true_matrix,cmap='viridis')
fig, ax = plt.subplots()
ax.set_title("Gemini Label")
im = ax.imshow(g_label,cmap='viridis')
fig, ax = plt.subplots()
ax.set_title("Gemini Scaled")
im = ax.imshow(g_scaled,cmap='viridis')
```
